```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mu1 = np.array([-3, 0])
mu2 = np.array([3, 0])

pi1 = 0.3
pi2 = 1-pi1

sigma = np.array([[2, 0.5],[0.5, 1]])

n = 500

x1, y1 = np.random.multivariate_normal(mu1, sigma, n).T
x2, y2 = np.random.multivariate_normal(mu2, sigma, n).T

choice1 = np.random.choice(2, n, p=[pi2, pi1])
choice2 = np.ones(n) - choice1

x = choice1 * x1 + choice2 * x2
y = choice1 * y1 + choice2 * y2

plt.plot(x, y, 'bx')
plt.axis('equal')

plt.show()
```
